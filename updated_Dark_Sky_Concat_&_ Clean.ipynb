```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
from sqlalchemy import Column, Integer, String, Float
import sqlalchemy
```

```python
dark_sky_csv1 = "mitchell_data/austin/austin_weather1.csv"
dark_sky_csv2 = "mitchell_data/austin/austin_weather2.csv"
```

```python
dark_sky_1 = pd.read_csv(dark_sky_csv1)
dark_sky_2 = pd.read_csv(dark_sky_csv2)
```

```python
frames = [dark_sky_1, dark_sky_2]
dark_sky_df = pd.concat(frames, ignore_index=True)
dark_sky_df.reset_index
dark_sky_df.head(2)
```

```python
len(dark_sky_df)
```

```python
dark_sky_df.drop_duplicates(subset=['Day'], keep= 'first', inplace = True)
len(dark_sky_df)
```

```python
dark_sky_df['Stan_Date'] = pd.to_datetime(dark_sky_df['Day'],unit='s')
dark_sky_df.head(1)
```

```python
dark_sky_df['Stan_Date'] = dark_sky_df['Stan_Date'].dt.date
dark_sky_df.head(1)
```

```python
dark_sky_df['Stan_Date'] = pd.to_datetime(dark_sky_df['Stan_Date'])
```

```python
dark_sky_df['WEEK'] = dark_sky_df['Stan_Date'].dt.week
dark_sky_df.head(2)
```

```python
dark_sky_df = dark_sky_df.reset_index()
```

```python
dark_sky_week_df = dark_sky_df.groupby(['WEEK']).mean()
```

```python
dark_sky_week_df.head(2)
```

```python
dark_sky_week_df = dark_sky_week_df.drop(columns = ['index','Day'])
dark_sky_week_df.head(2)
```
