```python
import pandas as pd
import numpy as np
```

```python
file = 'austin_output1.csv'
df = pd.read_csv(file)
df.head()
```

```python
df.columns
#detail, tempmax/min, humidity, wind speed,precipProbability, precipMax, visibility, sunrise, sunset 
```

```python
df1 = df.drop( columns = [ 'Moonphase', 'cloudCover', 'dewPoint', 'precipIntensityMax', 'pressure', 'temperatureMaxTime', 'temperatureMinTime', 'windGust', 'windGustTime'])
#df['Detail', 'temperatureMax', 'temperatureMin', 'windSpeed', 'precipProbability', 'precipIntensity', 'visibility', 'Sunrise', 'Sunset']
```

```python
df1['Day'] = pd.to_datetime(df1['Day'],unit='s')
df1['Sunrise'] = pd.to_datetime(df1['Sunrise'],unit='s')
df1['Sunset'] = pd.to_datetime(df1['Sunset'],unit='s')
```

```python
df1['YEAR'] = df1['Day'].dt.year
df1['MONTH'] = df1['Day'].dt.month
```

```python
df1 = df1.groupby('MONTH').mean()
```

```python
df1.to_csv('austin_weather.csv')
```
