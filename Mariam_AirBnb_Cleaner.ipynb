```python
import pandas as pd 
import numpy as np 
```

```python
df=pd.read_csv("austincalendar.csv")
df.head()
```

```python
df1 = df.drop(columns = ['available', 'minimum_nights', 'maximum_nights'])
```

```python
df1_group = df1.set_index('listing_id')
```

```python
df1_group['date'] = pd.to_datetime(df1_group['date'])
```

```python
df1_group['YEAR'] = df1_group['date'].dt.year
df1_group['MONTH'] = df1_group['date'].dt.month
df1_group['WEEK'] = df1_group['date'].dt.week
df1_group = df1_group.reset_index()
```

```python
df1_group.head()
```

```python
df1_group['price'].astype(str)
df1_group['price'] = df1_group.price.str.replace('$', '')
df1_group['price'] = df1_group.price.str.replace(',', '')
df1_group['price'] = df1_group['price'].astype(float)
```

```python
df1_group.dtypes
```

```python

df_week = df1_group.groupby(['WEEK']).mean()
```

```python
df_week
```

```python
df_week.reset_index
df_week.head()
```

```python
df_week = df_week.drop(columns = ['listing_id', 'MONTH', 'YEAR'])
```

```python
df_week['City'] = 'Austin'
df_week.head()
```

```python
df_week.to_csv('austin_clean.csv')
```
